```python
n,q=5, 1
S=[1, 1,1,1,1]
k=1401
# S=[1,2,3]
# k=7
def fibonacci(S,n):
    sum=0
    for i in S:
        sum+=i
    ln=n-len(S)-1
    return (sum*(2**ln))%(10**9+7)

 
fibonacci(S,k)
```
